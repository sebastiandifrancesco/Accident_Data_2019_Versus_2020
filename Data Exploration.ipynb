```python
# Count of 2020 Accidents for Each Divsion of US for purposes of investigating question above.
# This checks to make sure we have 10,000 data points for each division.
sev_sum = [pacific_df_2020['Severity'].count(),mountain_df_2020['Severity'].count(),west_north_central_df_2020['Severity'].count(),west_south_central_df_2020['Severity'].count(),east_north_central_df_2020['Severity'].count(),east_south_central_df_2020['Severity'].count(),new_england_df_2020['Severity'].count(),middle_atlantic_df_2020['Severity'].count(),south_atlantic_df_2020['Severity'].count()]
sev_sum
x_axis = np.arange(len(sev_sum))
plt.bar(x_axis, sev_sum, width = 0.8, color='r', alpha=0.5)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Pacific", "Mountain", "West North Central", "West South Central", "East North Central","East South Central",'New England','Middle Atlantic','South Atlantic'])
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(sev_sum)+5000)
plt.title('Count of 2019 Accidents for Each Divsion of US')
plt.xlabel('Divisions')
plt.ylabel('Count of Accidents')
plt.gcf().set_size_inches(20, 10)
```
