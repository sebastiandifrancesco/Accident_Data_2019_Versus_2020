```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
import os
```

```python
# Read in division files

accident_df_2019_enc = pd.read_csv('2019_east_north_central_division.csv')
accident_df_2019_esc = pd.read_csv('2019_east_south_central_division.csv')
accident_df_2019_xma = pd.read_csv('2019_middle_atlantic_division.csv')
accident_df_2019_xxm = pd.read_csv('2019_mountain_division.csv')
accident_df_2019_xne = pd.read_csv('2019_new_england_division.csv')
accident_df_2019_xxp = pd.read_csv('2019_pacific_division.csv')
accident_df_2019_xsa = pd.read_csv('2019_south_atlantic_division.csv')
accident_df_2019_wnc = pd.read_csv('2019_west_north_central_division.csv')
accident_df_2019_wsc = pd.read_csv('2019_west_south_central_division.csv')
#accident_df_2019_esc.head()
```

```python
# merge all Dataframes
frames = [accident_df_2019_enc,accident_df_2019_esc,accident_df_2019_xma,accident_df_2019_xxm,accident_df_2019_xne,
         accident_df_2019_xxp,accident_df_2019_xsa,accident_df_2019_wnc,accident_df_2019_wsc]
accident_df_2019_ttl = pd.concat(frames)
accident_df_2019_ttl.head()
```

```python
driver_stat = pd.read_csv('total_driver_age_state.csv')
driver_stat['State'] = driver_stat['State'].astype(str)
driver_stats = driver_stat.set_index('State')
driver_stats.head()
```

```python
#driver_stats.dtypes
```

```python
#accident_df_2019_enc.dtypes
```

```python
accident_data_w_demo = accident_df_2019_ttl.join(driver_stats,on='State')
accident_data_w_demo.head()
```

```python
accident_data_count = accident_data_w_demo.groupby('State').count()
```

```python
accident_data_sum = accident_data_w_demo.groupby('State').sum()
```

```python
accident_data_mean = accident_data_w_demo.groupby('State').mean()
```

```python
accident_data_merge = pd.merge(accident_data_count,accident_data_sum,on='State',suffixes=('_count','_sum'))
accident_data_merge.head()
```

```python
accident_data_merge2 = accident_data_merge.join(accident_data_mean,on='State')
print(accident_data_merge2.columns)
```

```python
column_1 = accident_data_merge2["Severity_sum"]
column_2 = accident_data_merge2["85+_sum"]
column_3 = accident_data_merge2["19_AND_UNDER_sum"]
column_4 = accident_data_merge2["Precipitation(in)"]
column_5 = accident_data_merge2["Temperature(F)"]
correlation_old = column_1.corr(column_2)
correlation_yng = column_1.corr(column_3)
correlation_pcp = column_1.corr(column_4)
correlation_tmp = column_1.corr(column_5)
print(correlation_old)
print(correlation_yng)
print(correlation_pcp)
print(correlation_tmp)
```
